```python
import pandas as pd

#Q1
#loading dataset
link = "https://raw.githubusercontent.com/MainakRepositor/Datasets/refs/heads/master/IPL/matches.csv"
data = pd.read_csv(link)

#display total no. of matches
total_matches = data.shape[0]
print("Total number of matches played in IPL history:", total_matches)

#print the column names
print("Columns in the dataset:", data.columns.tolist())

#first 5 rows of data
print(data.head(5))

#describe the data
print(data.describe())
```

```python
#Q2
#player of the match, matches won by just 1 run or 1 wicket decided on final ball
data = pd.read_csv("matches.csv")
dict_players = {}
for index, row in data.iterrows():
    if row['win_by_runs'] == 1 or row['win_by_wickets'] == 1:
        player_of_match = row['player_of_match']
        if player_of_match in dict_players:
            dict_players[player_of_match] += 1
        else:
            dict_players[player_of_match] = 1

#sorting the dictionary based on number of matches won
sorted_players = sorted(dict_players.items(), key=lambda x: x[1], reverse=True)
print("Players with most 'Player of the Match' awards in matches won by 1 run or 1 wicket decided on final ball:")
for player, count in sorted_players:
    print(f"{player}: {count} awards")
```

```python
#Q3
# At Wankhede Stadium, is it more common to win by batting first (runs) or by batting second
# (wickets)?

for index, rows in data.iterrows():
    if rows['venue'] == 'Wankhede Stadium':
        if rows['win_by_runs'] > 0:
            print("At Wankhede Stadium, it is more common to win by batting first (runs).")
        elif rows['win_by_wickets'] > 0:
            print("At Wankhede Stadium, it is more common to win by batting second (wickets).")
        break
```

```python
#Q4
# Which team has the highest number of wins where the victory margin was greater than 50 runs?

for index, rows in data.iterrows():
    if rows['win_by_runs'] > 50:
        print("The team with the highest number of wins where the victory margin was greater than 50 runs is:", rows['winner'])
        break

team_q4 = rows['winner']
```

```python
#Q5
# How many times has the team that won the toss also set a target and won the match?
count = 0
for index, rows in data.iterrows():
    if rows['toss_winner'] == rows['winner'] and rows['win_by_runs'] > 0 and rows['winner'] == team_q4: #team_q4 is delhi daredevils
        count += 1
print("The team that won the toss also set a target and won the match:", count, "times.")
```

```python
#Q6
# Which of umpire1 or umpire2 has officiated more matches involving KKR?
kkr_mask = (data['team1'] == 'Kolkata Knight Riders') | (data['team2'] == 'Kolkata Knight Riders')
kkr_matches = data[kkr_mask]

ump1_counts = kkr_matches['umpire1'].dropna().value_counts()
ump2_counts = kkr_matches['umpire2'].dropna().value_counts()

total_ump1 = ump1_counts.sum()
total_ump2 = ump2_counts.sum()

print("Total umpire1 appearances in KKR matches:", total_ump1)
print("Total umpire2 appearances in KKR matches:", total_ump2)

if total_ump1 > total_ump2:
    print("umpire1 has officiated more KKR matches.")
elif total_ump2 > total_ump1:
    print("umpire2 has officiated more KKR matches.")
else:
    print("umpire1 and umpire2 have officiated an equal number of KKR matches.")

# Optional: show top umpires in each column
print("\nTop umpire1s:\n", ump1_counts.head())
print("\nTop umpire2s:\n", ump2_counts.head())

```
